# Correlation Analysis

## Overview
This notebook analyzes the relationships between various features and startup success using correlation analysis. It loads a dataset of USA-based companies with success labels, visualizes the correlation matrix as a heatmap, and identifies the features most strongly correlated with the binary success indicator.

## Purpose
To identify which factors are most predictive of startup success by examining their statistical relationships, helping to inform feature selection for predictive modeling.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#correlation heatmap for USA companies with composite success labels
# Load your data
df = pd.read_csv('../4.define-success-metrics/usa_final_companies_with_success_labels.csv')

# Check the data looks good
print(df.head())

# Compute the correlation matrix (only numeric columns)
corr_matrix = df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, linewidths=.5)
plt.title('Correlation Matrix of Features and Target (USA Companies with Composite Success Labels)')
plt.tight_layout()  # Ensure the plot fits properly in the saved file
plt.savefig('correlation_heatmap_USA_composite.png', dpi=300)  # Save high-resolution plot
plt.show()

# Correlation of all features with the target (success_binary)
if 'success_binary' in df.columns:
    target_corr = corr_matrix['success_binary'].drop('success_binary')
    print("\nCorrelation of each feature with 'success_binary':\n")
    print(target_corr.sort_values(ascending=False))
else:
    print("'success_binary' column not found in your data.")
```
