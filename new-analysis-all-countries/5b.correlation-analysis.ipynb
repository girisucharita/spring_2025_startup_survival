```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your data
df = pd.read_csv('/Users/aminosaurier/Downloads/spring_2025_startup_survival/new-analysis-relationships-0/all_final_companies_with_success_labels.csv')

# Check the data looks good
print(df.head())

# Compute the correlation matrix (only numeric columns)
corr_matrix = df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, linewidths=.5)
plt.title('Correlation Matrix of Features and Target (success_binary)')
plt.show()

# Correlation of all features with the target (success_binary)
if 'success_binary' in df.columns:
    target_corr = corr_matrix['success_binary'].drop('success_binary')
    print("\nCorrelation of each feature with 'success_binary':\n")
    print(target_corr.sort_values(ascending=False))
else:
    print("'success_binary' column not found in your data.")
```
